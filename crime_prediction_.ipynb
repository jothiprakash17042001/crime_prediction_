```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import r2_score  # Import accuracy_score for classification

# Train the model
clf = LogisticRegression(random_state=0)
clf.fit(X_train, y_train)  # Use X_train and y_train for training

# Make predictions
y_pred = clf.predict(X_test)

# Evaluate the model using accuracy_score for classification
score = r2_score(y_test, y_pred)
print(score)
```
